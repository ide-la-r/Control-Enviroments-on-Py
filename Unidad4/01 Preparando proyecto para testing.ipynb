Vamos a preparar un proyecto simple al que añadiremos algunas pruebas automáticas mediante pytest. La estructura de directorios del proyecto será:


```
├── pytest.ini
├── requisitos.txt
├── src
│   └── calculos.py
└── tests
    └── test_calculos.py
```
En el directorio src incuiremos el código fuente de nuestro proyecto, los test estarán en el directorio tests, de esa forma diferenciamos completamente la parte de desarrollo de la de testing.


Para instalar los requisitos del proyecto vamos a crear un entorno virtual, lo activaremos e instalaremos todas las librerias necesarias. El fichero requisitos.txt contiene el listado de librerías a instalar

```
pytest
pytest-cov
```
Para crear el entorno e instalar las librerías ejecutamos en la raiz del proyecto:
```
python -m venv
Set-ExecutionPolicy -Scope CurrentUser -ExecutionPolicy RemoteSigned
.\.venv\Scripts\Activate.ps1  
python -m pip install --upgrade pip
pip install -r requisitos.txt
```
Para que pytest localice los ficheros de src el fichero debe contener lo siguiente:
```
[pytest]
pythonpath = src
```


Para poder probar sin pytest necesitamos crear una función main en cada módulo y ejecutar dentro las pruebas, pero estamos mezclando el código y las pruebas en el mismo fichero. A continuación podemos ver un ejemplo del fichero calculos.py

````python
def suma(a: int, b: int) -> int:
    return a + b

def divide(a: float, b: float) -> float:
    if b == 0:
        raise ValueError("No se puede dividir por 0")
    return a / b


def main():
    print(suma(2, 3))
    print(divide(10, 2))

if __name__ == "__main__":
    main()
````

````python
# Para escribir pruebas usamos el método assert que no hace nada si
# la parte izquierda de la comprobación conincide con la derecha
assert suma(2, 3) == 5
````

````python
# Vamos a probar un test que provoca un error
assert suma(2, 3) == 6
````

Nuestro fichero calculos.py debería quedar de la siguiente forma:
```
def suma(a: int, b: int) -> int:
    return a + b

def divide(a: float, b: float) -> float:
    if b == 0:
        raise ValueError("No se puede dividir por 0")
    return a / b
```
Ahora no contiene ninguna línea de testing

Nuestro fichero de pruebas src/test_calculos.py tendrá las pruebas


```
import pytest
from calculos import suma, divide

def test_suma():
    assert suma(2, 3) == 5

def test_divide_ok():
    assert divide(10, 2) == 5
```
para ejecutar los test escribimos en el terminal
```
pytest
```
